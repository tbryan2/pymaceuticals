## Pyamceuticals Analysis

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
%matplotlib inline

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df = pd.merge(mouse_metadata, study_results, how='outer')

# Display the data table for preview
df
```

```python
# Checking the number of mice.
#len(df['Mouse ID'].unique())
len(df)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
df_duplicates = df[df.duplicated(subset=['Mouse ID','Timepoint'])]

df_duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
df = df.drop_duplicates(['Mouse ID', 'Timepoint'])
```

```python
# Checking the number of mice in the clean DataFrame.
len(df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

summary_stats_mean = df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].mean()
summary_stats_mean = summary_stats_mean.rename(columns={'Tumor Volume (mm3)': 'Tumor Volume (mm3) Mean'})

summary_stats_median = df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].median()
summary_stats_median = summary_stats_median.rename(columns={'Tumor Volume (mm3)': 'Tumor Volume (mm3) Median'})

summary_stats_var = df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].var()
summary_stats_var = summary_stats_var.rename(columns={'Tumor Volume (mm3)': 'Tumor Volume (mm3) Variance'})

summary_stats_std = df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].std()
summary_stats_std = summary_stats_std.rename(columns={'Tumor Volume (mm3)': 'Tumor Volume (mm3) Std. Dev.'})

summary_stats_sem = df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].sem()
summary_stats_sem = summary_stats_sem.rename(columns={'Tumor Volume (mm3)': 'Tumor Volume (mm3) SEM'})

summary_stats_groupby = pd.concat([summary_stats_mean, summary_stats_median, 
                                  summary_stats_var, summary_stats_std, summary_stats_sem], axis=1)

summary_stats_groupby
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

summary_stats_agg = df[['Drug Regimen', 'Tumor Volume (mm3)']].groupby('Drug Regimen').agg(['mean', 'median', 'var', 'std', 'sem'])

summary_stats_agg
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
tp_per_regimen_count = df[['Drug Regimen', 'Timepoint']].groupby('Drug Regimen').count()


bar = tp_per_regimen_count.plot.bar(xlabel='Drug Regimen', ylabel='Count of Timepoints', legend=False, 
                                    title='Count of Timepoints per Drug Regimen')

bar;
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.
plt.bar(tp_per_regimen_count.index, tp_per_regimen_count['Timepoint'], align='center');
plt.title('Count of Timepoints per Drug Regimen')
plt.xlabel('Drug Regimen')
plt.ylabel('Count of Timepoints')
plt.xticks(rotation=90);
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
male_female_count = df[['Mouse ID', 'Sex']].groupby('Sex').count()

pie = male_female_count.plot.pie(y='Mouse ID', legend=False, title='Mice Count: Male v. Female')
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
plt.pie(male_female_count['Mouse ID'], labels=male_female_count.index);
plt.title('Mice Count: Male v. Female');
```
